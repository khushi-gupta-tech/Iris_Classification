Decision Tree Algorithm 

Decision Tree is a Supervised learning technique that can be used for both classification and Regression problems,but mostly it is preferred for solving Classification problems.It is a tree-structured classifier,where internodes represent the features of a dataset,branches represent the decision rules and each leaf node represents that outcome.

In a Decision tree,there are two nodes,which are the Decision Node and Leaf Node.Decision nodes are used to make any decision and have multiple branches,whereas Leaf nodes are the output of those decisions and do not contain any further branches.

The decisions or the test are performed on the basis of features of the given dataset.


Example-

Suppose there is a candidate who has a job offer and wants to decide whether he should accept the offer or Not.So,to solve this problem,the decision tree starts with the root node (Salary attribute by ASM). The root node splits further into the next decision node(distance from the office) and one leaf node based on the corresponding labels. The next decision node futher gets split into one decision node (Cab facility)
and one leaf node. Finally,the decision node splits into two leaf nodes (Accepted offers and Declined offer).

petal - short leaf

```python
# Import the essential Libraries
import numpy as np
import pandas as pd
```

```python
# Read the dataset
df=pd.read_csv("iris.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# separate dataset in independent and dependent variables
X=df.iloc[:,:4].values
X
```

```python
Y=df.iloc[:,4].values
Y
```

```python
# split dataset in training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

```python
# Apply Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

```python
# Test the model
y_pred = model.predict(X_test)
y_pred
```

```python
Y_test
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y_pred)
cm
```
